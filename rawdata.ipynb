## Raw Data
Create an annotations csv file with image filename and bounding box co-ordinates

```python
import json
import pandas as pd
```

```python
with open('probe_labels.json') as f:
  result = json.load(f)
  print(result)
data_train=[]
```

```python
for annotation in result['annotations']:
  image_id = annotation['image_id']
  bbox = annotation['bbox']

  for img in result["images"]:
    if img["id"] == image_id:
      file_name = img["file_name"]
      data_train.append({
         "id" : image_id,
         "file_name" : file_name,
         "bbox_coords" : bbox
      })
```

```python
df = pd.DataFrame(data_train)
df[['start_x','start_y','end_x','end_y']] = pd.DataFrame(df.bbox_coords.tolist(), index= df.index)
df = df.drop(['bbox_coords','id'], axis=1)
```

```python
df.to_csv('annotations.csv', index=False)
```
